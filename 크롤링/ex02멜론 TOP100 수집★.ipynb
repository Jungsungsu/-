```python
import requests as req
from bs4 import BeautifulSoup as bs
```

```python
req.get("https://www.melon.com/chart/index.htm")
```

```python
# 사람인척 속여서 데이터를 요청
h = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/96.0.4664.45 Safari/537.36'}
```

```python
# req를 통해서 MelonTop100데이터를 요청 . get()
# .get("요청할 페이지의 URL")
# 정상적일때는 [200] 응답코드 / 실패했을때는 [406]
# 헤더작업을 통해서 브라우저로 접속한 척 하기!
# get(headers = "헤드데이터")
res = req.get("https://www.melon.com/chart/index.htm", headers = h)
```

```python
# 응답코드속에 정확한 데이터가 담겨져있는지 확인
# .text 안에 있는 글자 정보를 출력을 해봐야함
res.text
```

```python
# bs을 통해서 컴퓨터가 알아먹을 수 있는 html, bs 객체화를 진행
# 이유는 res.text안에 있는 데이터가 문자열 데이터이기 때문에
soup = bs(res.text,"lxml")
```

```python
# 클래스값을 가지고 올때 중간에 공백이 들어가있는 경우 존재
# 공백(자손선택자)를 뜻하는 방법
# 클래스값이 여러개 있다! 공백을 지우고 .으로 이어주면 완성!
song = soup.select("div.ellipsis.rank01 > span")
```

```python
# 필요없는 개행문자(이스케이프코드) 지우는 방법
# strip() > 불필요한 글자들을 지워주는 메소드
song[0].text.strip()
```

```python
# 가수이름 수집시 a태그가 각각 여러개가 존재해서 개수가 다름
# 이런경우에는 a태그를 한번에 묶어주고있는 요소가 있는지 검사
# 있는경우는 그 묶음을 담당하는 요소를 수집
# 없는경우는 아쉽지만 수집후 직접 전처리가 필요!
singer = soup.select("div.ellipsis.rank02 > span.checkEllipsis")
```

```python
# 길이검사해서 정확한 데이터의 개수가 맞는지 검증
# 이 작업을 해야지만 밑에서 정확한 코딩이 가능!
# 귀찮더라도 습관적으로 꼭 확인하기!
len(song)
```

```python
len(singer)
```

### 선택자 가지고오는 방법
- 1. 요소의 클래스, 아이디값을 가진경우에는 직접 가지고오기
- 2. 클래스, 아이디가 없는 경우에는 부모태그를 검사
- 주의점. 부모태그를 가지고는 오는데 부모에게도 구분자가 없다면
  그 위에 있는 부모까지 검사를 진행!
- 부모를 검사했는데 데이터가 정확하지 않은경우!
  그 안에있는 다른요소를 검사
### 실제 개발에서 가장 많이 쓰이는 선택자
- 1순위 : class
- 2순위 : >(자식,자손 선택자)
- 3순위 :  nth-child(가상 선택자)
- 4순위 : id

# 데이터를 파일로 저장

```python
# 순수한 글자정보만 저장시켜주는 리스트 제작
# [리스트 제작 및 반복문 활용할때 팁]
# 리스트생성하는 코드와 반복문 돌리는 코드가 한셀안에 있으면 좋다
# 그 이유는 코드가 실행 될때마다 리스트가 매번 초기화가 되서
# 누적되는 오류가 발생하지 않음!
# 주의점! 반복문안에 리스트 생성하는 코드를 넣어버리면 매번 초기화가
# 진행되기 때문에 마지막 요소만 담기게 됨. 주의하기!
song_list = []
singer_list = []
rank_list = []
for i in range(len(song)) :
    # append 메소드 -> 차곡차곡 쌓아주기
    song_list.append(song[i].text.strip())
    singer_list.append(singer[i].text.strip())
    rank_list.append(i+1)
```

```python
rank_list
```

```python
len(singer_list)
```

# 표형태로 제작 후 파일로저장

```python
import pandas as pd
```

```python
# 컬럼, 데이터 구조를 만들기위해서 딕셔너리(dic)형태의 데이터를 제작
data = {"가수" : singer_list, "노래" : song_list, "순위" : rank_list}
```

```python
# 데이터를 표형태로 제작
melon = pd.DataFrame(data)
```

```python
# 인덱스 번호를 내가 원하는 컬럼으로 교체
# ,set_index("대체할컬럼")
melon.set_index("순위", inplace = True)
```

```python
melon
```

```python
# csv파일로 저장
# 1. euc-kr 인코딩
# 2. utf-8-sig 인코딩
# 3. ""

melon.to_csv("melon.csv", encoding ="euc-kr")
```
