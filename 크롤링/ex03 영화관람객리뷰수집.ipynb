```python
import requests as req
from bs4 import BeautifulSoup as bs
```

```python
h = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/96.0.4664.45 Safari/537.36'}
```

```python
res = req.get("https://movie.naver.com/movie/bi/mi/pointWriteFormList.naver?code=184311&type=after&isActualPointWriteExecute=false&isMileageSubscriptionAlready=false&isMileageSubscriptionReject=false", headers = h)
```

```python
soup = bs(res.text,'lxml')
```

```python
# 실제 리뷰데이터는 iframe 태그를 통해서 다른서버의 데이터를
# 빌려다가 보여주고만 있다.
# 데이터의 양이 너무 많기 때문에.
# iframe의 데이터를 수집하기 위해서 원본데이터의 주소를 찾기!
review = soup.select("div.score_reple > p")
```

```python
for i in review :
    print(i.text.strip())
```

```python
# 불필요한 관람객이라는 데이터가 등장
# 관람객 글자만 추출해서 지워버리기!
# 베스트방법은 선택자를 활용해서 정확히 가지고 오는거!
```

```python
#1. 관람객 글자만 수집
viewer = soup.select("span.ico_viewer")
```

```python
# 불필요한 요수(태그) 삭제하는 방법 extract()
# 원본데이터(soup)에 영향을 줘서 soup을 통해 데이터를 가지고 오는
# 모든 데이터에게 적용이 됨
for i in viewer :
    i.extract()
```

# 1페이지부터 10페이지까지 수집
- 변하는 url값을 찾아오는게 핵심
- url 맨뒤에 page=? 값을 기준으로 페이지가 변함

```python
# 반복문 작성할때는 항상 한번의 완벽한 로직을 먼저 작성
# 그 다음에 반복문으로 감싸주고 변해야하는 부분에 i만 넣어주면 반복문 작성
for i in tq(range(1,30)) :
    res = req.get("https://movie.naver.com/movie/bi/mi/pointWriteFormList.naver?code=184311&type=after&isActualPointWriteExecute=false&isMileageSubscriptionAlready=false&isMileageSubscriptionReject=false&page="+str(i))
    soup = bs(res.text,'lxml')
    viewer = soup.select("span.ico_viewer")
    for i in viewer :
        i.extract()
    review = soup.select("div.score_reple > p")
    for i in review :
        print(i.text.strip())
```

```python
# 반복의 진행상황을 체크해주는 라이브러리
# tq(range(?) 형식으로 range범위를 tq()로 감싸주면 반복상황 체크 가능)
from tqdm import tqdm_notebook as tq
```
