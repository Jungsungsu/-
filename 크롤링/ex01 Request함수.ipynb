# Requests 라이브러리
- 사용자가 원하는 페이지정보를 요청하기 위한 라이브러리

```python
# 라이브러리를 활용하기 위해서 임포트
import requests as req
```

```python
# req.get() => 원하는 페이지의 정보를 요청하는 메소드
# get("원하는 페이지의 URL")
# response [200] >> 데이터 통신에 성공했습니다!
res = req.get("http://www.naver.com")
```

```python
# 멜론사이트를 요청    400번대 에러는 요청 500번대는 서버나 DB 에러
# [406] >> 통신에 실패했습니다(문제점은 요청에서 문제가 있다.)
# 컴퓨터로 접속했다는 인지 (서버쪽에서)
req.get("https://www.melon.com")
```

```python
# 사람인척 속여서 데이터를 요청  (딕셔너리 형태)
h = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/96.0.4664.45 Safari/537.36'}
```

```python
# headers = "useragent정보를 같이 동반해서 요청"
# res는 result의 약자
res = req.get("https://www.melon.com", headers = h)
```

```python
res
```

```python
res.text
```

# BeautifulSoup 라이브러리 ★
- 가지고온 데이터에서 원하는 내용만 추출하게 도와주는 라이브러리
- 컴퓨터가 알아먹을 수 있는 (HTML 객체화를 시켜주는 라이브러리)

```python
# from bs4 >> bs4라는 공간으로부터 임포트를 하겠다.
# *주의점 : BeautifulSoup을 임포트! 스톤숩은 아님!
from bs4 import BeautifulSoup as bs
```

```python
#bs("바꿀데이터", "어떤방법 >> 파싱방법") ★
#컴퓨터가 접근할 수 있는 문서화(HTML)
soup = bs(res.text, "lxml")
```

```python
soup
```

```python
# 내가 필요한 블로그, 카페 등과 같은 요소를 들고오는 요소
# select("태그와 구분자(선택자)정보를 입력")
naver = soup.select("a.nav")
```

```python
naver
```

```python
# 데이터가 복수개이기 때문에 리턴타입 리스트타입
# 인덱스번호를 활용해서 원하는 요소만 출력
# 순수한 텍스트정보를 출력하기 위해서 .text입력
naver[0].text
```

```python
# 복수개를 다 출력할려면 반복문중에 for문 활용
for i in naver :
    print(i.text)
```
